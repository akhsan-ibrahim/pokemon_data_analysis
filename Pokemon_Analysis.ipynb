# **Analisis Data Pokemon**

```python
# Read CSV
import pandas as pd

df = pd.read_csv("Pokemon.csv")
df.head()
```

```python
# Information about DataFrame
df.info()
```

```python
# Description of DataFrame
df.describe()
```

```python
# Filtering DataFrame by Name, HP, and Speed column
df.filter(["Name","HP","Speed"])
```

```python
# Sorting based on Total column
df.sort_values("Total", ascending=True)
```

```python
# Group by Type 1 column
group_type1 = df.groupby("Type 1")
group_type1.first()
```

```python
# Aggregation (mean and median) of data Type 1 group
df.groupby("Type 1").agg(['mean','median'])
```

# Visualize

```python
import seaborn as sns
sns.scatterplot(data=df, x="Sp. Atk", y="Sp. Def", hue="Type 1")
```

```python
sns.lineplot(x="Sp. Atk", y="Sp. Def", data=df)
```

```python
sns.set(style="dark ")
sns.barplot(data=df, x="Type 1", y="Sp. Def")
```
